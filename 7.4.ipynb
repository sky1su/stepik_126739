Дан массив, состоящий из целых чисел. Напишите программу, подсчитывающую количество пар элементов массива, в которых оба числа двузначные. В данной задаче под парой подразумевается два подряд идущих элемента массива.

Входные данные:
Сначала задано число N — количество элементов в массиве (1 ≤ N ≤ 10000). Далее через пробел записаны N чисел — элементы массива. Элементы массива могут принимать целые значения от 0 до 10 000 включительно.

Выходные данные:
Необходимо вывести единственное число - количество пар элементов массива, в которых оба числа двузначные.

```python
# Считываем количество элементов в массиве
N = int(input())

# Считываем сам массив
array = list(map(int, input().split()))

# Проверяем, что количество элементов соответствует ожидаемому диапазону
assert 1 <= N <= 10000

# Инициализируем счетчик пар двузначных чисел
count_pairs = 0

# Перебираем все пары элементов в массиве
for i in range(N - 1):
    # Проверяем, являются ли текущий и следующий элементы двузначными
    if (array[i] >= 10 and array[i] <= 99) and (array[i + 1] >= 10 and array[i + 1] <= 99):
        count_pairs += 1

# Выводим результат
print(count_pairs)
```

# Задача №3
https://stepik.org/lesson/806256/step/4?unit=809385

Дан массив, состоящий из целых чисел. Напишите программу, которая подсчитает количество элементов массива, больших предыдущего (элемента с предыдущим номером).

Входные данные:
Сначала задано число N — количество элементов в массиве (1≤ N ≤10000). Далее через пробел записаны N чисел — элементы массива. Массив состоит из целых чисел.

Выходные данные:
Необходимо вывести единственное число - количество элементов массива, больших предыдущего.

```python
# Считываем количество элементов в массиве
N = int(input())

# Считываем элементы массива и преобразуем их в список целых чисел
array = list(map(int, input().split()))

# Проверяем, что количество элементов соответствует заданному диапазону
assert 1 <= N <= 10000

# Инициализируем счетчик элементов, больших предыдущего
count_greater = 0

# Перебираем все элементы массива, начиная со второго
for i in range(1, N):
    # Если текущий элемент больше предыдущего, увеличиваем счетчик
    if array[i] > array[i - 1]:
        count_greater += 1

# Выводим количество элементов, больших предыдущего
print(count_greater)
```

# Задача №4
https://stepik.org/lesson/806256/step/5?unit=809385

Дан массив, состоящий из целых чисел. Напишите программу, которая определяет, есть ли в массиве пара соседних элементов с одинаковыми знаками.

Входные данные:
Сначала задано число N — количество элементов в массиве (1 ≤ N ≤ 10000). Далее через пробел записаны N чисел — элементы массива. Массив состоит из целых чисел.

Выходные данные:
Необходимо вывести слово YES, если существует пара соседних элементов с одинаковыми знаками. В противном случае следует вывести слово NO.

```python
# Считываем количество элементов в массиве
N = int(input())

# Считываем элементы массива и преобразуем их в список целых чисел
array = list(map(int, input().split()))

# Проверяем, что количество элементов соответствует заданному диапазону
assert 1 <= N <= 10000

# Перебираем все пары соседних элементов в массиве
for i in range(1, N):
    # Проверяем, имеют ли соседние элементы одинаковые знаки
    if array[i] * array[i - 1] > 0:
        # Если да, выводим YES и завершаем программу
        print("YES")
        exit()

# Если все пары соседних элементов имеют разные знаки, выводим NO
print("NO")
```
