Юлий Цезарь использовал свой способ шифрования текста. Каждая буква заменялась на следующую по алфавиту через K позиций по кругу. Необходимо по заданной шифровке определить исходный текст.

Входные данные:
В первой строке дана шифровка, состоящая из заглавных латинских букв. Во второй строке число K (1 ≤ K ≤ 10).

Выходные данные:
Требуется вывести результат расшифровки.

```python
def decrypt(s: str, shift: int):
    start_ASCII = 65
    end_ASCII = 90
    result = ''
    k = 26
    for l in s:
        code = ord(l) - shift
        if start_ASCII <= code <= end_ASCII:   
            result += chr(code)
        if code < start_ASCII:
            result += chr(code + k)
        
    return result

s = input()
shift = int(input())

print(decrypt(s, shift))
```

# Задача №3
https://stepik.org/lesson/806252/step/3?unit=809381

В сети интернет каждому компьютеру присваивается четырехбайтовый код, который принято записывать в виде четырех чисел, каждое из которых может принимать значения от 0 до 255, разделенных точками. Вот примеры правильных IP-адресов:

127.0.0.0
192.168.0.1
255.0.255.255

Входные данные:
Программа получает на вход строку из ПРОИЗВОЛЬНЫХ символов.

Выходные данные:
Если эта строка является корректной записью IP-адреса, выведите YES, иначе выведите NO.

> проще было бы натравить проверку через регулярное выражение, но вы в них не умеете еще :)

```python
def check_ip(s: str):
    tmp = s.split('.')
    result = 'YES'
    if len(tmp) != 4:
        result = 'NO'
        return result
    for i in tmp:
        if i.isdigit() == False:
            result = 'NO'
            break
        if int(i) < 0 or int(i) > 255:
            result = 'NO'
            break
    return result

s = input()
print(check_ip(s))
```
