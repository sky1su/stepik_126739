# Задача №1
https://stepik.org/lesson/806237/step/2?unit=809366

В доме несколько подъездов. В каждом подъезде одинаковое количество квартир. Квартиры нумеруются подряд, начиная с единицы. Может ли в некотором подъезде первая квартира иметь номер x, а последняя – номер y?

Входные данныеВводятся два натуральных числа x и y (x ≤ y), не превышающие 10 000.

Выходные данные:
Выведите слово YES, если такое возможно, и NO в противном случае.

описание решения:
Данные вводятся произвольные, с соблюдением того, что это должны быть два натуральных числа, и x < y. Поэтому:

1. Для того, чтобы вычислить гипотетическое количество квартир в подъезде, нужно из номера последней квартиры вычесть количество квартир, которые должны находиться до первой квартиры 
```
y - (x - 1)
```
2. В доме должно быть целое количество подъездов, и в каждом подъезде одинаковое число квартир. Следовательно, количество квартир в предыдущих подъездах должно быть кратно количеству квартир в подъезде с заданными номерами.

````python
x, y = map(int, input().split())

if (x - 1) % (y - (x - 1)) == 0:
    print('YES')
else:
    print('NO')
````

# Задача №2
https://stepik.org/lesson/806237/step/3?unit=809366

Программа принимает три числа: возраст Антона, возраст Бориса и возраст Виктора. Определите, кто из них старше остальных.

Входные данные
Входная строка содержит три натуральных числа: возраст Антона, возраст Бориса и возраст Виктора.

Выходные данные
Выходная строка должна содержать латинскую букву – код того человека, который старше всех. Код Антона – 'A', код Бориса – 'B', код Виктора – 'C'. Если двое старше третьего, нужно вывести два кода через пробел (в алфавитном порядке). Например, если Антон и Виктор старше Бориса, программа должна вывести 'A C'. Если все трое одного возраста, программа должна вывести число 0.

````python
def get_result(data: dict):
    result = list()
    tmp_result = max(data, key=data.get)
    for key, value in data.items():
        if value == data[tmp_result]:
            result.append(key)
    if len(result) == 3:
        return 0
    return(' '.join(sorted(result)))
    

data = dict(zip(['A','B','C'], 
                map(int, input().split())
                )
            )

print(get_result(data))
````

# Задача №4
https://stepik.org/lesson/806237/step/4?unit=809366

Для заезда в оздоровительный лагерь организаторы решили заказать автобусы. Известно, что в лагерь собираются поехать N детей и M взрослых. Каждый автобус вмещает K человек. В каждом автобусе, в котором поедут дети, должно быть не менее двух взрослых.

Определите, удастся ли отправить в лагерь всех детей и взрослых, и если да, то какое минимальное количество автобусов требуется для этого заказать.

Входные данные
На вход программы поступают 3 натуральных числа, записанных через пробел - N, M и K, каждое из них не превосходит 10 000.

Выходные данные:
Выведите количество автобусов, которые нужно заказать. Если же отправить всех в лагерь невозможно, выведите число 0.

````python
from math import ceil as roundup
n, m, k = map(int, input().split())
count_passagers = n + m
count_bus = roundup(count_passagers / k)
count_parrents = m // count_bus

if count_parrents >= 2:
    print(count_bus)
else:
    print(0)
````
